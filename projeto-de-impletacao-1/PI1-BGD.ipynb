### Questão 2
Liste as 10 palavras que mais ocorrem na coleção


```python
import re
rgx = re.compile(r'(.+)\t(\d+)\n')
def getWordMap(collection):
  with open(f'output/{collection}/part-00000', 'r') as f:
    map = {}
    lines = f.readlines()
    for line in lines:
      result = rgx.search(line)
      map[result.group(1)] = int(result.group(2))
    return {k: v for k, v in sorted(map.items(), key=lambda item: item[1], reverse=True)}

def showTop10():
  collections = ['shakespeare','gnutella']
  for collection in collections:
    map = getWordMap(collection)
    top10 = list(map.items())[:10]
    print(f'** Top 10 palavras da coleção {collection} **')
    for word, rank in top10:
      print(f'palavra: {word}     quantidade: {rank}')
    print('-'*50)

showTop10()
```

### Questão 3
Apresente um histograma da frequência de ocorrência das palavras na coleção. O histograma deverá ter 10 classes, cada uma representando um intervalo de frequência. O formato do histograma deverá ser textual, e não gráfico. Por exemplo, seja Ii o í-ésimo intervalo de frequência e Cn número de palavras cuja frequência de ocorrência está no intervalo, o histograma teria o seguinte formato:

I1 - C1  
I2 - C2  
In - Cn  

```python
import numpy as np

def showHistogram():
  collections = ['shakespeare','gnutella']
  for collection in collections:
    map = getWordMap(collection)
    hist, bins = np.histogram(list(map.values()), range = (0.0, 520.0))
    print(f'** Histograma da coleção {collection} **')
    for i in range(len(hist)):
      print(f'l{i+1}[{bins[i]}, {bins[i+1]}]    C{i+1} = {hist[i]}')
    print('-'*50)

showHistogram()
```
